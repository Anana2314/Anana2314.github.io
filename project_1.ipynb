This project explores the World Bank Population dataset, focus especially in 2023 to later make a gender population analysis. 

```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasets/population/master/data/population.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.tail()
```

We have explore the dataset and know calculate the mean, media and mode of population for the year 2023

```python
# Filtrar solo el aÃ±o 2023
df_2023 = df[df["Year"] == 2023]

# Calcular medidas
mean_val = df_2023["Value"].mean()
median_val = df_2023["Value"].median()
mode_val = df_2023["Value"].mode().iloc[0]

print("Mean population:", round(mean_val, 2))
print("Median population:", round(median_val, 2))
print("Mode population:", mode_val)
```

```python

mean_val = statistics.mean(pop_2023)
median_val = statistics.median(pop_2023)

try:
    mode_val = statistics.mode(pop_2023)
except statistics.StatisticsError:
    mode_val = statistics.multimode(pop_2023)[0]


print("Mean population (2023):", round(mean_val, 2))
print("Median population (2023):", round(median_val, 2))
print("Mode population (2023):", mode_val)
```

As a last exercise, we want to see the total population by income group

```python

with open("API_SP.POP.TOTL_DS2_en_csv_v2_130083.csv", newline="", encoding="utf-8") as f:
    for _ in range(4):
        next(f)  # skip metadata rows
    reader = csv.DictReader(f)

    income_map = {
        "High income": ["United States", "Germany", "Japan", "United Kingdom", "France", "Canada", "Australia"],
        "Upper middle income": ["China", "Russia", "Brazil", "Mexico", "South Africa", "Turkey"],
        "Lower middle income": ["India", "Indonesia", "Pakistan", "Bangladesh", "Nigeria", "Kenya"],
        "Low income": ["Ethiopia", "Sudan", "Mozambique", "Nepal", "Madagascar"]
    }

    income_pop = {k: 0 for k in income_map.keys()}

    for row in reader:
        country = row["Country Name"]
        value = row.get("2020")
        if not value or value.strip() == "":
            continue
        try:
            pop = float(value)
        except ValueError:
            continue

        for group, countries in income_map.items():
            if country in countries:
                income_pop[group] += pop
                break

print("\nðŸ’° Total Population by Income Group (2020)\n")
max_val = max(income_pop.values())
scale = 50 / max_val

for group, pop in sorted(income_pop.items(), key=lambda x: x[1], reverse=True):
    bar = "â–ˆ" * int(pop * scale)
    print(f"{group:<20} | {bar} {int(pop/1_000_000):,}M")
```
