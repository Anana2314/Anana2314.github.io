We have explore the dataset and the next step is to explore population in 2024 which can serve us to later make comparisions respect to other years. We will calculate the mean, media and mode of adult population for the year 2024. The column selected is pop_adult

```python
#First filter only 2024
df_2024 = df[df["year"] == 2024]

#Calculate mean, median, mode
mean_val = df_2024["pop_adult"].mean()
median_val = df_2024["pop_adult"].median()
mode_val = df_2024["pop_adult"].mode().iloc[0]

print("Mean population:", round(mean_val, 2))
print("Median population:", round(median_val, 2))
print("Mode population:", mode_val)
```

```python
# We already have the list of value so we to calculate the mean we do the basic mean = sum / count

mean_value = sum(values) / len(values)
print("Mean:", mean_value)
```

```python
# To calculate de median, we sorted the values
values_sorted = sorted(values)
n = len(values_sorted)
middle = n // 2

if n % 2 == 1:
    # if the number is odd: then the median is the middle one
    median_value = values_sorted[middle]
else:
    # if the list is even number: average of the two middle numbers
    median_value = (values_sorted[middle - 1] + values_sorted[middle]) / 2
print("Median:", median_value)
```

As a last exercise, we calculate the mean population of Argentina for the years 2017, 2021, 2024. Following this example https://computing-in-context.afeld.me/project_1.html we created a graph with the mean population for those years. 

```python
import pandas as pd

#We used the same findex dataset 
url = "/Users/anamontanez/Documents/Anana2314.github.io/PROJECT 1/GlobalFindexDatabase2025.csv"
df = pd.read_csv(url)

#  Selected only Argentina and the data from the last three more recent surveys
pais = "Argentina"
df_arg = df[
    (df["countrynewwb"] == pais) &
    (df["year"].isin([2017, 2021, 2024]))
].copy()

# There are different values for each year, that's why we used the mean
df_arg_year = (
    df_arg.groupby("year", as_index=False)["pop_adult"]
    .mean()
)

# Scale of 10 characters to make the graph easy to read
max_val = df_arg_year["pop_adult"].max()
scale = 10 / max_val if max_val > 0 else 1

print("\nArgentinian Population (2017, 2021, 2024)\n")

# Graph sparkline, 
for _, row in df_arg_year.iterrows():
    year = row["year"]
    pop = row["pop_adult"]
    bar_len = int(pop * scale)
    bar = "**" * bar_len
    print(f"{year}: {bar} ({int(pop):,})")
```
